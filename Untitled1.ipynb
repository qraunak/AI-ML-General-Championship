```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('xx.csv')

# Remove leading and trailing whitespaces from column names
df.columns = df.columns.str.strip()

# Print DataFrame information for debugging
print("Columns in DataFrame:")
print(df.columns)

# Print a list of columns (excluding the 'image' column) for debugging
print("List of columns (excluding 'image'):")
print(list(df.columns)[1:])

# Iterate through columns (excluding the 'image' column) and create separate CSV files
for column in df.columns[1:]:
    print(f"Processing column: {column}")
    # Create a DataFrame for each column where the value is 1
    subset_df = df[df[column] == 1]
    
    # Save the subset DataFrame to a new CSV file
    subset_df.to_csv(f'{column}_output.csv', index=False)

print("Processing completed.")
```

```python
import os
import shutil

# Path to the folder containing the images
folder_path = r"C:\Users\raush\Desktop\temp\Kaggle GC\aiml-general-championship\KCDH2024_Training_Input_10K\KCDH2024_Training_Input_10K"

# Path to the CSV file
csv_file_path = r"C:\Users\raush\Desktop\temp\delete2\files\MEL_output.csv"

# Read the image names from the CSV file
with open(csv_file_path, 'r') as file:
    image_names = [line.strip() for line in file.readlines()]

# Create a new folder to store the selected images
new_folder_path = r"C:\Users\raush\Desktop\temp\delete2\files\MEL"
os.makedirs(new_folder_path, exist_ok=True)

# Copy the selected images to the new folder
for image_name in image_names[1:]:
    print(image_name[:12])
    source_path = os.path.join(folder_path, f"{image_name[:12]}.jpg")  # Assuming the images have the '.jpg' extension
    destination_path = os.path.join(new_folder_path, f"{image_name[:12]}.jpg")
    
    # Check if the source file exists before copying
    if os.path.exists(source_path):
        shutil.copy2(source_path, destination_path)
        print(f"Image {image_name}.jpg copied successfully.")

print("Image copy process completed.")
```
